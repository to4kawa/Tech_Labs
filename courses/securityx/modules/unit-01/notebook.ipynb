# Unit 01 補助ノート

目的: Cloud Shellで取得した失敗/成功ログを整理し、Explain-backを作る。

```python
# 失敗/成功ログを貼り付けて比較
failed = {
    'principal': 'learner@example.com',
    'granted': False,
    'status': 'Permission iam.serviceAccounts.update denied'
}
success = {
    'principal': 'learner@example.com',
    'granted': True,
    'status': ''
}
print('FAILED:', failed)
print('SUCCESS:', success)
```

```python
# Explain-back下書き生成
template = f"""
1分説明:
- 何をしたか: IAM最小権限で更新操作を試し、失敗後に必要権限のみ追加して再検証。
- なぜ必要か: 過剰権限を避けつつ運用要件を満たすため。
- 何が証拠か: granted={failed['granted']}（失敗）→ granted={success['granted']}（成功）。
"""
print(template)
```
